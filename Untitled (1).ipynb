```python
import plotly.graph_objs as go
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df_ = px.data.gapminder()
```

```python
df_
```

```python
#group by year and continent
df=df_.groupby(['year','continent']).sum().reset_index()
df.head()
```

```python
import plotly.graph_objects as go


fig = go.Figure(
    data=[
        go.Bar(
            name='Asia',
            x=df[df.continent == 'Asia']['year'],
            y=df[df.continent == 'Asia']['pop']
        ),
        go.Bar(
            name='Europe',
            x=df[df.continent == 'Europe']['year'],
            y=df[df.continent == 'Europe']['pop']
        )
    ]
)
fig.show()
```

```python
country1= 'United States'
country2='Japan'
```

```python
fig = go.Figure()
```

```python
fig.add_trace(
    go.Scatter(
        x=df_[df_.country==country1].year,
        y=df_[df_.country==country1].lifeExp,
        name=country1,
        text=df_[df_.country==country1].lifeExp,
        mode= 'lines+text', 
        textposition='bottom right'
    )
)

fig.add_trace(
    go.Scatter(
        x=df_[df_.country==country2].year,
        y=df_[df_.country==country2].lifeExp,
        name=country2,
        mode='lines',
        line=dict(
            color='sienna',
            width=5
        )
    )
)


fig.show()
```

```python
fig.update_layout(
    margin=dict(
        l=5,
        r=5,
        t=50,
        b=5
    )
)
fig.show()
```

```python
df = pd.read_csv('2010YumaAZ.csv')
days = ['TUESDAY','WEDNESDAY','THURSDAY','FRIDAY','SATURDAY','SUNDAY','MONDAY']
```

```python
df
```

```python
data = []

for day in days:
    trace = go.Scatter(
        x=df['LST_TIME'],
        y=df[df['DAY']==day]['T_HR_AVG'],
        mode='lines',
        name=day
    )

    data.append(trace)

layout = go.Layout(
    title='Daily temperatures from June 1-7, 2010 in Yuma, Arizona',
    hovermode='closest'
)

fig = go.Figure(data=data, layout=layout)
```

```python
fig.show()
```

```python
a_df = pd.read_csv('abalone.csv')
```

```python
a_df
```

```python
fig = go.Figure(
    data= [
        go.Histogram(
            x=a_df[(a_df['length']>=0) & (a_df['length']<=1)]['length'],
            xbins=dict(
                size=0.02
            )
        )
    ]
)

fig.show()

fig.update_layout(
    title='Shell lengths from the Abalone dataset',
    xaxis_title='length'
)
```

```python
df = pd.read_csv('mpg.csv')
```

```python
df
```

```python
import plotly.graph_objects as go

data = [
    go.Scatter(
        x=df['displacement'],
        y=df['acceleration'],
        mode='markers',
        marker=dict(color='blue'),
        customdata=df['name'],  # Attach the name column to each point
        hovertemplate=
            'Displacement: %{x}<br>' +
            'Acceleration: %{y}<br>' +
            'Name: %{customdata}<extra></extra>'
    )
]

layout = go.Layout(
    title='Vehicle acceleration vs displacement',
    yaxis_title='Acceleration = seconds to reach 60mph',
    xaxis_title='Displacement',
    height=500,
    hovermode='closest'
)

fig = go.Figure(
    data=data,
    layout=layout
)

fig.show()
```
